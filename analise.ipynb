```python
# Análise da Variável Alvo (SalePrice)

sns.histplot(df['SalePrice'], kde=True, bins=30)
plt.title("Distribuição de SalePrice")
plt.show()
```

```python
#calcular assimetria (Skewness)
from scipy.stats import skew

skewness = skew(df['SalePrice'])
print(f"Skewness: {skewness:.2f}")

# |skew| < 0.5 → aproximadamente normal
# 0.5 ≤ |skew| < 1 → moderadamente assimétrica
# |skew| ≥ 1 → fortemente assimétrica
```

```python
#Como SalePrice é fortemente assimétrica é necessário aplicar o log para aproximar os dados

df['SalePrice_log'] = np.log(df['SalePrice'])

sns.histplot(df['SalePrice_log'], kde=True, bins=30)
plt.title("Distribuição de SalePrice após log")
plt.show()

print(f"Skewness após log: {skew(df['SalePrice_log']):.2f}")

df.drop(columns=['SalePrice_log']);
```

```python
# Correlação com Features Numéricas

colunasNumericas = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = colunasNumericas.corr()

top_corr = corr_matrix['SalePrice'].abs().sort_values(ascending=False)[1:11]

print("Top 10 features numéricas mais correlacionadas com SalePrice:")
print(top_corr)


top_features = top_corr.index.tolist() + ['SalePrice']  # incluir SalePrice para referência
plt.figure(figsize=(12,6))
sns.heatmap(df[top_features].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap das Top 15 Features Numéricas mais correlacionadas com SalePrice")
plt.show()
```

```python
# Scatterplot (Gráfico de Dispersão)

top_features = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath']

plt.figure(figsize=(15,10))

for i, feature in enumerate(top_features):
    plt.subplot(2, 3, i+1)  # organiza múltiplos gráficos
    sns.scatterplot(x=df[feature], y=df['SalePrice'])
    plt.title(f'SalePrice vs {feature}')
    plt.xlabel(feature)
    plt.ylabel('SalePrice')

plt.tight_layout()
plt.show()
```

```python
def detectar_outliers_iqr(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 3 * IQR
    limite_superior = Q3 + 3 * IQR
    
    return df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]

# Exemplo de uso
outliers = detectar_outliers_iqr(df, 'GrLivArea')
print(outliers[['GrLivArea','SalePrice']])
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['GrLivArea'], y=df['SalePrice'], label='Casas Normais')
sns.scatterplot(x=outliers['GrLivArea'], y=outliers['SalePrice'], color='red', label='Outliers')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.title('Detecção de Outliers em GrLivArea vs SalePrice')
plt.legend()
plt.show()
```
