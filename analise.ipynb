```python
#calcular assimetria (Skewness)
from scipy.stats import skew

skewness = skew(df['SalePrice'])
print(f"Skewness: {skewness:.2f}")

# |skew| < 0.5 â†’ aproximadamente normal
# 0.5 â‰¤ |skew| < 1 â†’ moderadamente assimÃ©trica
# |skew| â‰¥ 1 â†’ fortemente assimÃ©trica
```

```python
#Como SalePrice Ã© fortemente assimÃ©trica Ã© necessÃ¡rio aplicar o log para aproximar os dados

df['SalePrice_log'] = np.log(df['SalePrice'])

sns.histplot(df['SalePrice_log'], kde=True, bins=30)
plt.title("DistribuiÃ§Ã£o de SalePrice apÃ³s log")
plt.show()

print(f"Skewness apÃ³s log: {skew(df['SalePrice_log']):.2f}")

df.drop(columns=['SalePrice_log']);
```

```python
# CorrelaÃ§Ã£o com Features NumÃ©ricas

colunasNumericas = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = colunasNumericas.corr()

top_corr = corr_matrix['SalePrice'].abs().sort_values(ascending=False)[1:11]

print("Top 10 features numÃ©ricas mais correlacionadas com SalePrice:")
print(top_corr)


top_features = top_corr.index.tolist() + ['SalePrice']  # incluir SalePrice para referÃªncia
plt.figure(figsize=(12,6))
sns.heatmap(df[top_features].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap das Top 15 Features NumÃ©ricas mais correlacionadas com SalePrice")
plt.show()
```

```python
def detectar_outliers_iqr(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 3 * IQR
    limite_superior = Q3 + 3 * IQR
    
    return df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]

# Exemplo de uso
outliers = detectar_outliers_iqr(df, 'GrLivArea')
print(outliers[['GrLivArea','SalePrice']])
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['GrLivArea'], y=df['SalePrice'], label='Casas Normais')
sns.scatterplot(x=outliers['GrLivArea'], y=outliers['SalePrice'], color='red', label='Outliers')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.title('DetecÃ§Ã£o de Outliers em GrLivArea vs SalePrice')
plt.legend()
plt.show()
```
